```python
import matplotlib.pyplot as plt
from astropy.io import fits
import numpy as np
from scipy.ndimage import gaussian_filter1d
from scipy.signal import medfilt
```

```python
with fits.open("spEigenGal-53054.fits") as hdul:
    # print(hdul.info())
    data_full = hdul[0].data  # continumm + lines data
    spec = data_full.astype(np.float64)
    data_continumm = medfilt(spec[0], kernel_size=301)
    # data_continumm = gaussian_filter1d(data_full, sigma=150)
    data_lines = data_full[0] - data_continumm


plt.plot(data_lines)
plt.show()
```

```python
# plotting spectra from a data file of galaxy - spec-1678-53433-0425.fits

with fits.open("spec-1678-53433-0425.fits") as hdul:
    spec = hdul[1].data

flux = spec["flux"]
wavelength = 10**spec["loglam"]
ivar = spec["ivar"]

import matplotlib.pyplot as plt

plt.plot(wavelength, flux)
plt.xlabel("Wavelength (Å)")
plt.ylabel("Flux")
plt.show()
```
